```python
# 19. Add values on top of bars
import matplotlib.pyplot as plt

products = ["A", "B", "C"]
sales = [50, 80, 60]

bars = plt.bar(products, sales)

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height(),
             str(bar.get_height()),
             ha='center')

plt.show()
```

```python
# 40. Dashboard
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot([1,2,3],[2,4,6])
plt.title("Line")

plt.subplot(2,2,2)
plt.bar(["A","B","C"], [5,7,4])
plt.title("Bar")

plt.subplot(2,2,3)
plt.hist(np.random.randn(100))
plt.title("Histogram")

plt.subplot(2,2,4)
plt.scatter([1,2,3],[4,5,6])
plt.title("Scatter")

plt.tight_layout()
plt.show()
```

```python
# 43. Seaborn scatter plot
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
# 48. Seaborn box plot
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.boxplot(data=tips, x="day", y="total_bill")
plt.show()
```

```python
# 49. Heatmap using correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

corr = tips.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.show()
```

```python
# 50. Titanic dataset - three visualizations
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# Visualization 1
sns.countplot(data=titanic, x="sex")
plt.show()

# Visualization 2
sns.histplot(data=titanic, x="age")
plt.show()

# Visualization 3
sns.boxplot(data=titanic, x="class", y="fare")
plt.show()
```
